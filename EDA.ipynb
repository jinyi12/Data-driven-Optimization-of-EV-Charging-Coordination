```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import datetime
```

```python
# read output.csv
df = pd.read_csv('output.csv')
```

```python
# remove blank space in column names
df.columns = df.columns.str.replace(' ', '')
```

```python
# group df by i and get each group
df_group = df.groupby('i')
df_group = [df_group.get_group(x) for x in df_group.groups]
```

```python
#  read arrival_departure data
df_arrival_departure = pd.read_csv('Data/arrival_departure_times.csv')
```

```python
# round off each time to integer
df_arrival_departure['arrival'] = df_arrival_departure['arrival'].apply(lambda x: int(x))
df_arrival_departure['departure'] = df_arrival_departure['departure'].apply(lambda x: int(x))
```

```python
#  recalculate the duration
df_arrival_departure['duration'] = df_arrival_departure['departure'] - df_arrival_departure['arrival']
```

```python
# get the arrival and departure time of first behicle
df_arrival_departure.iloc[0]["arrival"]
```

```python
# for each group, test if DayAheadChargingPower is all zeros, append the result to a list
# and test if the list contains False
DayAheadChargingPower_all_zeros = []
for i in range(len(df_group)):
    if df_group[i]['DayAheadChargingPower'].sum() == 0:
        DayAheadChargingPower_all_zeros.append(True)
    else:
        DayAheadChargingPower_all_zeros.append(False)

# if the list contains False, print the index of the group
if False in DayAheadChargingPower_all_zeros:
    print([i for i, x in enumerate(DayAheadChargingPower_all_zeros) if x == False])
```
